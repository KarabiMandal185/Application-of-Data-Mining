```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/employee_performance.csv')

# Display the first few rows of the dataset
print(df.head())
```

```python
# Generate statistical summaries
summary = df.describe()
print(summary)
```

```python
# Data Transformation
# Create a new column for Salary to Experience Ratio
df['Salary_to_Experience'] = df['Monthly_Salary'] / (df['Years_of_Experience'] + 1)  # Adding 1 to avoid division by zero

# Display the updated DataFrame
print(df.head())
```

```python
import matplotlib.pyplot as plt

# Histogram of Monthly Salaries
plt.figure(figsize=(10, 6))
plt.hist(df['Monthly_Salary'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Monthly Salaries')
plt.xlabel('Monthly Salary')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
import seaborn as sns

# Scatter Plot: Years of Experience vs. Performance Score
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Years_of_Experience', y='Performance_Score', hue='Department', style='Department', s=100)
plt.title('Years of Experience vs. Performance Score')
plt.xlabel('Years of Experience')
plt.ylabel('Performance Score')
plt.grid()
plt.show()
```

```python
# Heatmap of numerical features
plt.figure(figsize=(10, 6))
correlation_matrix = df[['Monthly_Salary', 'Years_of_Experience', 'Performance_Score']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap of Numerical Features')
plt.show()
```
